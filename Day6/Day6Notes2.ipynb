## Boston Datasets

```python
data = load_boston()
dir(data)
```

```python
data['data']
```

```python
data = load_boston(return_X_y=True)
data[0][15]
```

```python
data = load_boston()
df = pd.DataFrame(data['data'], columns=data['feature_names'])
df['Harga'] = data['target']
df.tail()
```

```python
# sb.heatmap(df.corr(), annot=True)
df.corr()
```

```python
from sklearn import preprocessing
```

```python
dfx = df.drop('Harga', axis=1)
dfy = pd.DataFrame(df['Harga'])
```

```python
dfxS = pd.DataFrame(preprocessing.scale(dfx),
                   columns=data['feature_names'])
dfxS['Harga'] = dfy
dfxS.corr()
```

```python
# splitting 90% training, 10% testing
x = dfxS[['RM', 'PTRATIO', 'LSTAT']]
y = dfxS['Harga']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y,
    train_size = .9
)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
hargaPred = model.predict(x_test)
harga = y_test
```

```python
model.score(x_test, y_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(harga, hargaPred)
rmse = np.sqrt(mse)
rmse
```

```python
r2 = r2_score(harga, hargaPred)
r2
```

## Iris Datasets

```python
from sklearn.datasets import load_iris
```

```python
i = load_iris()
dir(i)
```

```python
i['target_names']
```

```python
df = pd.DataFrame(i['data'], columns=['SL', 'SW', 'PL', 'PW'])
df['spesies'] = i['target']
df['namaSp'] = df['spesies'].apply(
    lambda x: i['target_names'][x]
)
df
```

```python
df0 = df[df['spesies'] == 0]
df1 = df[df['spesies'] == 1]
df2 = df[df['spesies'] == 2]
```

```python
plt.figure(figsize=(12, 6))
plt.suptitle('Iris Dataset', fontsize=23)

plt.subplot(121)
plt.plot(df0['SL'], df0['SW'], 'ro')
plt.plot(df1['SL'], df1['SW'], 'go')
plt.plot(df2['SL'], df2['SW'], 'yo')
plt.legend(['setosa', 'versicolor', 'virginica'])
plt.title('Sepal length vs Sepal width')
plt.xlabel('Sepal length (cm)')
plt.ylabel('Sepal width (cm)')
plt.grid(True)

plt.subplot(122)
plt.plot(df0['PL'], df0['PW'], 'ro')
plt.plot(df1['PL'], df1['PW'], 'go')
plt.plot(df2['PL'], df2['PW'], 'yo')
plt.legend(['setosa', 'versicolor', 'virginica'])
plt.title('Petal length vs Petal width')
plt.xlabel('Petal length (cm)')
plt.ylabel('Petal width (cm)')
plt.grid(True)

plt.show()
```
