```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('insurance.csv')
df = df[['age', 'bmi', 'children', 'charges']]
df
```

```python
scaler = StandardScaler()
z = scaler.fit_transform(df[['age', 'bmi', 'children', 'charges']])
z
```

```python
df['z age'] = z[:,0]
df['z bmi'] = z[:,1]
df['z children'] = z[:,2]
df['z charges'] = z[:,3]
df
```

```python
plt.subplot(231)
plt.scatter(df['age'], df['charges'])
plt.subplot(232)
plt.scatter(df['bmi'], df['charges'])
plt.subplot(233)
plt.scatter(df['children'], df['charges'])

plt.subplot(234)
plt.scatter(df['z age'], df['z charges'])
plt.subplot(235)
plt.scatter(df['z bmi'], df['z charges'])
plt.subplot(236)
plt.scatter(df['z children'], df['z charges'])

plt.show()
```

```python
model = LinearRegression()
model.fit(df[['z age', 'z bmi', 'z children']], df['z charges'])
m = model.coef_
c = model.intercept_
print(m)
print(c)
```

```python
df['z charges Å·'] = model.predict(df[['z age', 'z bmi', 'z children']])
df
```
