### Anime Recommendation

__2. Collaburative filtering__

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('rating.csv')
df = df.iloc[:5000]
df.head()
```

```python
df.isnull().sum()
```

```python
dfNew = df.pivot_table(index='user_id', columns='anime_id', aggfunc='mean')
dfNew.head()
```

```python
dfCor = dfNew.corr()
dfCor = dfCor.loc['rating']
dfCor.loc[1]
```

```python
# index, rating
MrA = [(1, 5), (5, 5), (6, 6)]
MrA
```

```python
skorSama = pd.DataFrame()
for index, rating in MrA:
    skor = dfCor.iloc[rating]
    skor = skor.sort_values(ascending=False)
    skorSama = skorSama.append(skor, ignore_index=True)

skorSama
```

```python
skorSama.sum().sort_values(ascending=False)
```
