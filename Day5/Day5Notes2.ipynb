```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_excel('Normalisasi.xlsx', header=1)
df = df[['TB']]
df
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, MaxAbsScaler, RobustScaler, Normalizer, QuantileTransformer, Binarizer
```

```python
# Standardisasi / z-score normalisasi
df['Z Std'] = StandardScaler().fit_transform(df[['TB']])

# normalisasi / min max scaler
df['MinMax'] = MinMaxScaler().fit_transform(df[['TB']])

# max abs scaler
df['MaxAbs'] = MaxAbsScaler().fit_transform(df[['TB']])

# Robust scaler
df['Robust'] = RobustScaler().fit_transform(df[['TB']])

# Quantile Transformer normal dist / gaussian
df['Gaussian'] = QuantileTransformer(output_distribution='normal').fit_transform(df[['TB']])

# Quantile Transformer uniform dist
df['Uniform'] = QuantileTransformer(output_distribution='uniform').fit_transform(df[['TB']])

# Normalizer
df['Normalizer'] = Normalizer().fit_transform(df[['TB']])

# Binarizer
df['Binarizer'] = Binarizer(threshold=0.85).fit_transform(df[['TB']])

df
```

```python
sb.pairplot(df)
```

```python
# inverse Z-std => nilai awal
scz = StandardScaler().fit(df[['TB']])
scz.inverse_transform(df[['Z Std']])

# inverse MinMax => nilai awal
mms = MinMaxScaler().fit(df[['TB']])
mms.inverse_transform(df[['MinMax']])

# inverse MaxAbs => nilai awal
mas = MaxAbsScaler().fit(df[['TB']])
mas.inverse_transform(df[['MaxAbs']])

# inverse Robust => nilai awal
robust = RobustScaler().fit(df[['TB']])
robust.inverse_transform(df[['Robust']])
```
