```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
df= pd.read_csv('insurance.csv')
df
```

```python
label = LabelEncoder()
dfUji = df
dfUji['sex'] = label.fit_transform(dfUji['sex'])
dfUji['smoker'] = label.fit_transform(dfUji['smoker'])
dfUji['region'] = label.fit_transform(dfUji['region'])
dfUji
```

```python
x = dfUji[['sex', 'smoker', 'region', 'age', 'bmi', 'children']]
y = dfUji['charges']
x
```

```python
ct = ColumnTransformer(
    [('one_hot_encoder', OneHotEncoder(categories='auto'), [0, 1, 2])
    ],
    remainder='passthrough'
)

x = np.array(ct.fit_transform(x), dtype=np.int64)
x
```

```python
model = LinearRegression()
model.fit(x, y)
```

```python
df['charges_pred'] = model.predict(x)
df
```
