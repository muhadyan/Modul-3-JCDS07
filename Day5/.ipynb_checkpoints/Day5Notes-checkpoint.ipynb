```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Normalisasi.xlsx', header=1)
df = df[['TB', 'BB']]
df
```

<hr>

### 1. Normalisasi

- Normalisasi $\displaystyle x' = \frac {x - x_{min}} {x_{max} - x_{min}} $

```python
df['nTB 1'] = (df['TB'] - df['TB'].min()) / (df['TB'].max() - df['TB'].min())
df['nBB 1'] = (df['BB'] - df['BB'].min()) / (df['BB'].max() - df['BB'].min())
df
```

<hr>

### 2. Normalisasi Sklearn minmax_scale

```python
from sklearn.preprocessing import minmax_scale
```

```python
df['nTB 2'] = minmax_scale(df['TB'])
df['nBB 2'] = minmax_scale(df['BB'])
df
```

<hr>

### 3. Normalisasi MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['TB', 'BB']])
n = scaler.transform(df[['TB', 'BB']])
df['nTB 3'] = n[:, 0]
df['nBB 3'] = n[:, 1]
df
```

```python
scaler.inverse_transform(
    df[['nTB 1', 'nBB 1']]
)
```
